```python
import pandas as pd
import numpy as np
from restaurantKey import apiKey, clientKey
import pprint as pprint
```

```python
openTableURL = " http://opentable.herokuapp.com/api/restaurants/zip:"
yelpURL = "https://api.yelp.com/v3/businesses/search"
```

```python
responseList1 = []
responseList2 = []
```

```python
zipCodes = ("./zipCodes.csv")
zipDF = pd.read_csv(zipCodes)
zipDF = zipDF.drop_duplicates(keep='first', inplace=False)
zipDF
```

```python
zipDF.dtypes
```

```python
zipList = zipDF.values.tolist()
print(zipList)
```

```python
zipList2 = []
for zip in zipDF['Zip code']:
    zipList2.append(zip)
    
print(zipList2)
```

```python
for zip in zipList2:
    print("Testing:\t"+str(zip))
```

```python
headers = {'Authorization': 'Bearer %s' % apiKey}

for zip in zipList2:
    params = {'category':'restaurants,All','location':f'{zip}'}

    # Making a get request to the API
    req=requests.get(yelpURL, params=params, headers=headers)
    # proceed only if the status code is 200
    print('The status code is {}'.format(req.status_code))
    responseList1.append(req.json())

# printing the text from the response 
print(responseList1)

#imported from postmaster
#https://api.yelp.com/v3/businesses/search?category=restaurants, All&location=60625, 60640, 60631, 60630, 60618, 60630, 60610, 60625, 60613, 60625, 60631, 60634, 60640, 60645, 60660, 60614, 60618, 60641, 60659
```
